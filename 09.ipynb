```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_path = 'C:/Users/trand/img/hun.jpg'
im = Image.open(my_path)
plt.imshow(im)
```

```python
img_gray = im.convert('L')
plt.imshow(img_gray, cmap = 'gray')
```

```python
im_array = np.array(img_gray)
print(im_array)
```

```python
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biểu đồ Histogram của ảnh')
plt.xlabel('giá trị của pixel')
plt.ylabel('số lượng')
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(im_array, origin="image")
plt.show()
```
